```python
df=pd.read_csv("car_purchasing.csv",encoding='latin-1')
df.head()
```

```python
import matplotlib.pyplot as plt
dim=["age","annual Salary","credit card debt","net worth"]
corr_matrix = df[dim].corr()
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap - car purchasing")
plt.show()
```

```python
df=df.drop(columns=['customer name', 'customer e-mail', 'country'],axis=1)
```

```python
X = df.iloc[:, :-1].values
Y = df.iloc[:,-1].values
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#using decition tree regressor
from sklearn.tree import DecisionTreeRegressor
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
clf = DecisionTreeRegressor(random_state=0)
clf.fit(X_train,Y_train)
```

```python
Y_pred = clf.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```
