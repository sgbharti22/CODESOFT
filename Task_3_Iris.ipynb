```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dim = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
ds[dim].boxplot()
plt.xlabel("Dimensions")
plt.ylabel("Values")
plt.title("Box Plots - Iris Dataset")
plt.show()
```

```python
import seaborn as sns
corr_matrix = ds[dim].corr()
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation Heatmap - Iris Dataset")
plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(25, 6), sharey=True)
sns.histplot(ds, ax=axes[0], x="sepal_length", kde=True)
sns.histplot(ds, ax=axes[1], x="sepal_width", kde=True)
sns.histplot(ds, ax=axes[2], x="petal_length", kde=True)
sns.histplot(ds, ax=axes[3], x="petal_width", kde=True)
```

```python
ds.replace({'species':{'Iris-setosa' : 0 ,'Iris-versicolor': 1, 'Iris-virginica' : 2}},inplace=True)
```

```python
ds
```

```python
data=ds.values
X=data[:,0:4]
Y=data[:,4]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(Y_test, Y_pred)
print("confusion matrix : ")
print(cm)
print("Accuracy:")
print(accuracy_score(Y_test, Y_pred))
print("classification report : ")
print(classification_report(Y_test, Y_pred))
```
